```python
df.columns = df.columns.str.strip()
```

```python
df["Grade"].unique()
df["Gender"].unique()
df["Department"].unique()
df["Extracurricular_Activities"].unique()
df["Internet_Access_at_Home"].unique()
df["Parent_Education_Level"].unique()
df["Family_Income_Level"].unique()
```

```python
df["Grade"] = df["Grade"].map({'A': 0, 'B': 1,'C':2,'D':3,'F':4})
df["Gender"]=df["Gender"].map({'Male':0,'Female':1})
df["Department"]=df["Department"].map({'Engineering':0,'Business':1,'Mathematics':2,'CS':3})
df["Extracurricular_Activities"]=df["Extracurricular_Activities"].map({'No':0,'Yes':1})
df["Internet_Access_at_Home"]=df["Internet_Access_at_Home"].map({'No':0,'Yes':1})
df["Parent_Education_Level"]=df["Parent_Education_Level"].map({'High School':0,"Bachelor's":1,"Master's":2,'PhD':3})
df["Family_Income_Level"]=df["Family_Income_Level"].map({'Low':0,'Medium':1,'High':2})
```

```python
df.head()
```

```python
x = df.drop(["Grade"], axis = 1)
y = df["Grade"]
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Define a dictionary of models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'Naive Bayes': GaussianNB()
}

# Store results for each model
results = []

# Loop through the models, train, predict and evaluate
for model_name, model in models.items():
    # Step 1: Train the model
    model.fit(x_train, y_train)
    
    # Step 2: Make predictions
    y_pred = model.predict(x_test)
    
    # Step 3: Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average=None)
    recall = recall_score(y_test, y_pred, average=None)
    f1 = f1_score(y_test, y_pred, average=None)
    classification_rep = classification_report(y_test, y_pred, output_dict=True)
    
    # Store the results in the results list
    results.append({
        "models": model_name,
        "accuracy": accuracy,
        "precision": precision,  # Convert numpy array to list if needed
        "recall": recall,        # Convert numpy array to list if needed
        "f1_score": f1          # Convert numpy array to list if needed
    })

# Convert the results list to a DataFrame
metrics_score = pd.DataFrame(results)

# Round the values to 2 decimal places
metrics_score = metrics_score.reset_index(drop=True)

# Optionally, print the DataFrame
print(metrics_score)
```

```python
x = df.drop(["Stress_Level (1-10)"], axis = 1)
y = df["Stress_Level (1-10)"]
```

```python
x.columns
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
# Import necessary libraries for regression
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define a dictionary of regression models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Support Vector Regression': SVR(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Decision Tree': DecisionTreeRegressor(max_depth=4)
}

# Store results for each model
results = []

# Loop through the models, train, predict and evaluate
for model_name, model in models.items():
    # Step 1: Train the model
    model.fit(x_train, y_train)
    
    # Step 2: Make predictions
    y_pred = model.predict(x_test)
    
    # Step 3: Evaluate the model
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    
    # Store the results in the results list
    results.append({
        "model": model_name,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R-squared": r2
    })

# Convert the results list to a DataFrame
metrics_score = pd.DataFrame(results)

# Round the values to 2 decimal places
metrics_score = metrics_score.round(2)

# Optionally, print the DataFrame
print(metrics_score)
```
